```python
import osmnx as ox
import pandas as pd
from shapely.geometry import Point

def get_universities_and_amenities(cities, amenities, dist=500):
    data = []
    
    for city in cities:
        # Query for universities
        universities = ox.geometries_from_place(city, tags={"amenity":"university"})
        print(f"Found {len(universities)} universities in {city}")

        for _, university in universities.iterrows():
            # University name and coordinates
            uni_name = university['name'] if 'name' in university else 'Unknown'
            
            if isinstance(university['geometry'], Point):
                uni_coords = (university['geometry'].y, university['geometry'].x)
            else:  # if it's a Polygon or MultiPolygon
                uni_coords = (university['geometry'].centroid.y, university['geometry'].centroid.x)

            # For each amenity, get places around university
            for amenity in amenities:
                places_around = ox.geometries_from_point(center_point=uni_coords, tags={"amenity":amenity}, dist=dist)
                print(f"Found {len(places_around)} {amenity} places around {uni_name} in {city}")

                # If there are any places found, add them to the data list
                for _, place in places_around.iterrows():
                    place_name = place['name'] if 'name' in place else 'Unknown'
                    data.append({'City': city, 'UniversityName': uni_name, 'Category': amenity, 'Name': place_name})
                    
    # Create a DataFrame from the data list
    df = pd.DataFrame(data)
    return df

cities = ["Istanbul, Turkey", "Ankara, Turkey", "Izmir, Turkey"]
amenities = ["restaurant", "cafe", "coffee_shop"]
df = get_universities_and_amenities(cities, amenities)
```
